```python
import pickle
def read_dict_file(file_path):
    with open(file_path, 'rb') as f:
            data = pickle.load(f)
    return data
```

```python
text_dict1 = read_dict_file('【高阶】《运行手册》HO-0907 飞行运行.pkl')
text_dict2 = read_dict_file('大型飞机公共航空运输承运人运行合格审定规则.pkl')
text_dict2_lv2 = read_dict_file('大型飞机公共航空运输承运人运行合格审定规则 lv2.pkl')
```

```python
import pandas as pd
df = pd.read_excel('banchmark.xlsx')
test_data = df['文本']

for item in test_data:
    print(item)
    print('\n')
```

```python
import jieba

# 加载停用词表
def load_stopwords(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        stopwords = set(line.strip() for line in f)
    stopwords.add(' ')
    return stopwords

# 分词并去除停用词
def segment_and_remove_stopwords(text, stopwords):
    words = jieba.lcut(text)
    filtered_words = [word for word in words if word not in stopwords]
    return filtered_words


stopwords_file = "stopwords.txt"
stopwords = load_stopwords(stopwords_file)

processed_test_data = []
for item in test_data:
  i = segment_and_remove_stopwords(item, stopwords)
  processed_test_data.append(i)
```

```python
def process_dict(input_dict):
    output_dict = {}
    # 对字典中的每一个 value 进行处理
    for key, value in input_dict.items():
        # 确保 value 是字符串类型
        if isinstance(value, str):
            output_dict[key] = segment_and_remove_stopwords(value,stopwords)
    return output_dict
processed_dict2_lv2 = process_dict(text_dict2_lv2)
```
