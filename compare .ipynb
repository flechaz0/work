```python
import pandas as pd

data_file_path = "Benchmark.xlsx"
data = pd.read_excel(data_file_path)
diff_contents = []

for index,row in data.iterrows():
    r7_text = row['r7_text_content']
    r8_text = row['r8_text_content']
    diff_content = get_diff(r7_text,r8_text)
    diff_contents.append(diff_content)
    
data['diff1'] = diff_contents
data.to_excel(data_file_path, index=False)
```
